## NLP

Understand basic NLP concepts and apply text preprocessing techniques.

```python
import pandas as pd

data = {
    "text": [
        "I love machine learning!",
        "NLP is amazing and fun",
        "I hate bugs in code",
        "Debugging is very frustrating sometimes",
        "I enjoy learning new technologies"
    ]
}

df = pd.DataFrame(data)
df
```

```python
df["lower"] = df["text"].str.lower()
df
```

**Tokenization**

```python
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize

df["tokens"] = df["lower"].apply(word_tokenize)
df
```

**Remove Stopwords**

```python
nltk.download('stopwords')

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

df["no_stopwords"] = df["tokens"].apply(
    lambda words: [word for word in words if word not in stop_words]
)

df
```

**Stemming**

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

df["stemmed"] = df["no_stopwords"].apply(
    lambda words: [stemmer.stem(word) for word in words]
)

df
```

**Lemmatization**

```python
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

df["lemmatized"] = df["no_stopwords"].apply(
    lambda words: [lemmatizer.lemmatize(word) for word in words]
)

df
```

```python
df["clean_text"] = df["lemmatized"].apply(lambda words: " ".join(words))
df
```
