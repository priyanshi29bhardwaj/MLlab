## GenAI

Understand the fundamentals of Generative AI and experiment with prompt-based tools. 

```python
import pandas as pd

df = pd.read_csv("loan_data.csv")
df.head()
```

Prompt 1 – Dataset Summary

```python
def genai_tool(prompt, data):
    
    if "summary" in prompt.lower():
        return "This dataset contains loan details such as credit policy, purpose, interest rate, and repayment status."
    
    elif "insight" in prompt.lower():
        return """1. Higher interest rates may increase default chances.
2. Credit policy approval affects loan success.
3. Loan purpose influences repayment behavior."""
    
    elif "feature" in prompt.lower():
        return "Important features include interest rate, income, credit score, and loan purpose."
    
    elif "generate" in prompt.lower():
        return "Loan analysis shows patterns in borrower behavior and financial risk."
    
    else:
        return "Provide a clear prompt."
```

```python
sample_data = df.head(5).to_string()

print("🔹 Prompt 1: Summary")
print(genai_tool("Give summary of dataset", sample_data))

print("\n🔹 Prompt 2: Insights")
print(genai_tool("Give insights", sample_data))

print("\n🔹 Prompt 3: Feature Analysis")
print(genai_tool("Explain features", sample_data))

print("\n🔹 Prompt 4: Text Generation")
print(genai_tool("Generate analysis", sample_data))
```

```python
def ai_prediction(row):
    if row['int.rate'] > 0.12:
        return "High Risk"
    else:
        return "Low Risk"

df['AI_Risk'] = df.apply(ai_prediction, axis=1)
df.head()
```
