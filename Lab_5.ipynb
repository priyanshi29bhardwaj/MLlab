## DESCRIPTIVE STATISTICS

```python
import pandas as pd

df = pd.read_excel("new_data.xlsx")

# First look
print(df.head())
```

```python
# Structure
print(df.info())
```

```python
# Statistical summary
print(df.describe())
```

Features: X1–X8
Targets available: Y1 and Y2 (both continuous)

Since logistic regression is for classification, we’ll first convert one target into a binary class.

## Logistic Regression

we create a Target for Logistic Regression

We convert Y1 into two classes:

1 = High energy building (above median)

0 = Low energy building (below median)

```python
# Create binary target from Y1
df['Target'] = (df['Y1'] > df['Y1'].median()).astype(int)

print(df['Target'].value_counts())
```

```python
##Data Cleaning: this datset is already clean
print(df.isnull().sum())
```

```python
print(df.duplicated().sum())
```

```python
## Feature Selection
X = df[['X1','X2','X3','X4','X5','X6','X7','X8']] ##independent variables
y = df['Target'] ## target variable
```

```python
##Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
##Train-Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
## Logistic Regression
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
### Convert categorical columns to category type
df['X6'] = df['X6'].astype('category')
df['X8'] = df['X8'].astype('category')

# One-hot encoding
df = pd.get_dummies(df, columns=['X6','X8'], drop_first=True)

print(df.head())
```

```python
## model evaluation
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:,1]

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_prob))
```

## USA HOUSING DATASET

```python
import pandas as pd

df = pd.read_csv("USA_Housing.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
## data cleaning
print(df.isnull().sum())
```

```python
print(df.duplicated().sum())
```

```python
df = df.drop(columns=['Address'])  # Address is text, not useful
```

```python
## descriptive statistics
df.describe()
```

This dataset has no categorical columns, so:

✅ No label encoding
✅ No one-hot encoding needed

(All columns are already numeric)

We convert price into binary categories:

1 → Expensive House (above median price)

0 → Affordable House (below median price)

```python
df['Price_Class'] = (df['Price'] > df['Price'].median()).astype(int)

print(df[['Price','Price_Class']].head())
```

```python
X = df.drop(columns=['Price','Price_Class'])
y = df['Price_Class']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:,1]

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_prob))
```
